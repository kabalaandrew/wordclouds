```python
from os import path
from scipy.misc import imread
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
from wordcloud import WordCloud, STOPWORDS
import sys
%matplotlib inline

##additional parameters
# rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Computer Modern Roman']
rcParams['text.usetex'] = False
rcParams['figure.figsize'] = 25, 15
```

```python
d = path.dirname('__file__')

# Read the whole text.
text = open(path.join(d, 'alice.txt')).read()

# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
#alice_mask = np.array(Image.open(path.join(d, "alice_mask.png")))

alice_mask = imread("./alice_mask.png", flatten=True)
wc = WordCloud(background_color="black", font_path="C:\\Windows\\Fonts\\Verdana.ttf", max_words=2000, mask=alice_mask,
               stopwords=STOPWORDS.add("said"), width=2000, height=500)

# generate word cloud
wc.generate(text)

# store to file
#wc.to_file(path.join(d, "alice.png"))

# show
plt.imshow(wc)
plt.axis("off")
plt.figure()
#plt.imshow(alice_mask, cmap=plt.cm.gray)
plt.axis("off")
plt.savefig('Alice_cloud.png', dpi=300)
#plt.show()
```
